# Auditor√≠a 4: Estudio de Variables Externas

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import datetime as dt
import matplotlib.dates as mdates

from pyspark import SparkContext, SQLContext
from pyspark.sql import SQLContext
from pyspark.sql.types import *

sqlCtx = SQLContext(sc)
```

### 1. Carga de los Datos

```python
df_general_linea14=sqlContext.sql("select fecha,avg(n_viajeros) as n_viajeros,avg(inten_lluvia) as inten_lluvia_media,avg(ocupacion_trafico) as inten_trafico_medio,avg(intensidad_evento) as inten_evento_medio from emt_smartbus.viajeros_tramos_master_nooutliers where linea=14 GROUP BY fecha")
df_general_lineaN22=sqlContext.sql("select fecha,avg(n_viajeros) as n_viajeros,avg(inten_lluvia) as inten_lluvia_media,avg(ocupacion_trafico) as inten_trafico_medio,avg(intensidad_evento) as inten_evento_medio from emt_smartbus.viajeros_tramos_master_nooutliers where linea=522 GROUP BY fecha")
```

### 2. Estudio Variables

#### 2.1. Plot Linea 14

```python
f1 = [int(i.fecha) for i in df_general_linea14.select('fecha').collect()]
f1_sort=sorted(f1)
m1= map(str,f1_sort)
avg_nviajeros1 = [float(i.n_viajeros) for i in df_general_linea14.select('n_viajeros').collect()]
avg_inten_lluvia1 = [i.inten_lluvia_media for i in df_general_linea14.select('inten_lluvia_media').collect()]
avg_inten_trafico1 = [i.inten_trafico_medio for i in df_general_linea14.select('inten_trafico_medio').collect()]
avg_inten_evento1 = [i.inten_evento_medio for i in df_general_linea14.select('inten_evento_medio').collect()]
```

```python
# Intensidad de viajeros
x = [dt.datetime.strptime(d,'%Y%m%d').date() for d in m1]
plt.figure(figsize=(10,5))
plt.figure(1)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y%m%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.plot(x,avg_nviajeros1)
plt.gcf().autofmt_xdate()
plt.title('Intensidad media de viajeros para la linea 14')
plt.xlabel('fecha')
plt.ylabel('intensidad media de viajeros')

# Intensidad de Lluvia
plt.figure(figsize=(10,5))
plt.figure(2)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y%m%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.plot(x,avg_inten_lluvia1)
plt.gcf().autofmt_xdate()
plt.title('Intensidad media de la Lluvia para la linea 14')
plt.xlabel('fecha')
plt.ylabel('intensidad lluvia media')

#Intensidad de Trafico
plt.figure(figsize=(10,5))
plt.figure(3)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y%m%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.plot(x,avg_inten_trafico1)
plt.gcf().autofmt_xdate()
plt.title('Intensidad media del trafico para la linea 14')
plt.xlabel('fecha')
plt.ylabel('intensidad trafico medio')

#Intensidad de Eventos
plt.figure(figsize=(10,5))
plt.figure(4)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y%m%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.plot(x,avg_inten_evento1)
plt.gcf().autofmt_xdate()
plt.title('Intensidad media de los eventos para la linea 14')
plt.xlabel('fecha')
plt.ylabel('intensidad eventos medios')
```
